## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#This chart is to show which song genres are the most listened to based on the data, which will give an idea about what song genres to use when creating business ads to market company products

df['Fav genre'].hist(bins=35, figsize=(25, 10), color = ['purple'])
plt.xlabel('Most Popular Genres', fontsize=15)
plt.ylabel('Number of Users', fontsize=15)
```

```python
# #This chart is to show the number of people and their described music effects from the three most popular genres (Rock, Pop, Metal) 

df3 = df[['Fav genre', 'Music effects']]
df3.groupby('Fav genre').value_counts() 
df4 = df[(df['Fav genre'] == 'Rock') | (df['Fav genre'] == 'Pop') | (df['Fav genre'] == 'Metal')]

# #Seaborn plot 
df5 = df4.groupby('Music effects')['Fav genre'].value_counts().unstack().plot.bar(stacked=True)
df5.set_title('Music Effects of Most Popular Genres')
df5.set_ylabel('Number of Users')
```
